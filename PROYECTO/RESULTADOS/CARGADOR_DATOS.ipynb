### Carga de librerias

```python
import pandas as pd
import sys
import matplotlib.pyplot as plt
```

### Cargar daata

```python
df= pd.read_csv("../DATA/Algerian_forest_fires_dataset_CLEANED.csv", sep=",")
df.head
```

```python
sys.path.append('../DATA')
from CARGA_ANALISIS import cargador_datos
```

```python
 # Número de filas y columnas
filas = df.shape[0]
columnas = df.shape[1]
```

### Graficos

```python
# Gráfico de barras para mostrar la cantidad de filas y columnas
plt.figure(figsize=(6, 6))
plt.bar(['Filas', 'Columnas'], [filas, columnas], color=['skyblue', 'lightcoral'])
plt.title('Cantidad de Filas y Columnas', fontsize=14)
plt.ylabel('Cantidad')
plt.show()
```

```python
# Gráfico de pie para mostrar el porcentaje de nulos vs no nulos
nulos = df.isnull().sum().sum()
no_nulos = filas * columnas - nulos
plt.figure(figsize=(6, 6))
plt.pie([nulos, no_nulos], labels=['Nulos', 'No Nulos'], autopct='%1.1f%%', startangle=90, colors=['red', 'green'])
plt.title('Porcentaje de Nulos vs No Nulos', fontsize=14)
plt.show()
```

```python
# Gráfico de barras para mostrar los tipos de variables (cuántas columnas de cada tipo)
tipos_variables = df.dtypes.value_counts()
plt.figure(figsize=(8, 6))
tipos_variables.plot(kind='bar', color='purple')
plt.title('Tipos de Variables en el Dataset', fontsize=14)
plt.xlabel('Tipo de Variable')
plt.ylabel('Cantidad de Columnas')
plt.xticks(rotation=45)
plt.show()
```
